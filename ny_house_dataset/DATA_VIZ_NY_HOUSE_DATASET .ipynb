```python
### IMPORTING NECESSARY PACKAGES
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
import os
os.startfile(os.getcwd())
```

```python
### LOADING DATASET
data = pd.read_csv(r"C:\Users\HP\Documents\NY-House-Dataset.csv")
#pd.set_option('display.max_rows', None)
data
```

```python
### REMOVING UNNECESSARY COLUMNS
del data['MAIN_ADDRESS']
del data['ADMINISTRATIVE_AREA_LEVEL_2']
del data['LOCALITY']
del data['STREET_NAME']
del data['LONG_NAME']
del data['FORMATTED_ADDRESS']
data
```

```python
### CONVERTING FROM ONE DATATYPE TO ANOTHER
data['BATH'] = data['BATH'].astype('int64')
data['BATH']
```

```python
sns.countplot(data, x = 'TYPE')
plt.xticks(rotation= 'vertical')
plt.title('COUNT OF HOUSING TYPES')
plt.show()
```

```python
data['STATE'] = data['STATE'].apply(lambda x: x.split(', ')[0])
sns.countplot(data, x = 'STATE', width= 1.0)
plt.xticks(rotation= 'vertical')
plt.title('COUNT OF HOUSES PER STATE')
plt.tight_layout()
plt.show()
```

```python
plt.plot(data['BEDS'], data['BATH'], 'x')
plt.show()
```

```python
sns.countplot(data, x = 'BEDS')
plt.show()
```

```python
brooklyn_houses = data['TYPE'].loc[(data['STATE'] == 'Brooklyn')]
count = brooklyn_houses.value_counts()
plt.pie(count, labels= count.index)
plt.show()
```
